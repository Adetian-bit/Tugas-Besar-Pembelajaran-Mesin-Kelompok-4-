### Step 5 - Membuat Target Klasifikasi
Cell ini mengubah target numerik `Paddy yield(in Kg)` menjadi dua kelas. Nilai di atas median masuk kategori High Yield, sedangkan nilai sama dengan atau di bawah median masuk kategori Low Yield.

```python
target_col = 'Paddy yield(in Kg)'
threshold = df[target_col].median()

df['Yield_Category'] = (df[target_col] > threshold).astype(int)

df['Yield_Label'] = df['Yield_Category'].map({
    0: 'Low Yield',
    1: 'High Yield'
})

print("Threshold median:", threshold)
print("\nAturan klasifikasi:")
print("Low Yield  = Paddy yield <=", threshold)
print("High Yield = Paddy yield >", threshold)

print("\nDistribusi kelas:")
display(df['Yield_Label'].value_counts())

print("\nDistribusi kelas dalam persen:")
display((df['Yield_Label'].value_counts(normalize=True) * 100).round(2))
```

Distribusi kelas menjadi cukup seimbang, yaitu **1.171 Low Yield** dan **1.167 High Yield**. Ini baik untuk klasifikasi karena tidak ada kelas yang terlalu dominan.

### Step 7 - Memisahkan Fitur dan Target
Cell ini memisahkan variabel input dan target. Kolom target asli tidak dimasukkan ke fitur agar tidak terjadi data leakage.

```python
X = df.drop(columns=[target_col, 'Yield_Category', 'Yield_Label'])
y = df['Yield_Category']

categorical_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Jumlah data:", X.shape[0])
print("Jumlah fitur sebelum encoding:", X.shape[1])

print("\nFitur kategorik:")
print(categorical_cols)

print("\nFitur numerik:")
print(numeric_cols)

print("\nTarget:")
print("Yield_Category, dengan 0 = Low Yield dan 1 = High Yield")
```

### Penjelasan Output
Output menunjukkan jumlah data setelah preprocessing adalah **2.338 baris** dengan **44 fitur** sebelum encoding. Program memisahkan fitur kategorik dan fitur numerik. Fitur kategorik seperti `Agriblock`, `Variety`, `Soil Types`, dan arah angin akan diubah dengan One Hot Encoding. Fitur numerik seperti luas lahan, pupuk, suhu, curah hujan, dan kelembapan langsung digunakan sebagai input model. Target yang dipakai adalah `Yield_Category`, yaitu 0 untuk Low Yield dan 1 untuk High Yield.

### Step 8 - Membagi Data Training dan Testing
Cell ini membagi data menjadi 80 persen data training dan 20 persen data testing. Stratifikasi digunakan agar proporsi kelas tetap seimbang.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Ukuran X_train:", X_train.shape)
print("Ukuran X_test :", X_test.shape)

print("\nDistribusi target pada data training:")
display((y_train.value_counts(normalize=True) * 100).round(2))

print("\nDistribusi target pada data testing:")
display((y_test.value_counts(normalize=True) * 100).round(2))
```

### Penjelasan Output
Output menunjukkan data dibagi menjadi data training dan testing. Data training berisi **1.870 baris**, sedangkan data testing berisi **468 baris**. Distribusi target pada training dan testing tetap seimbang karena memakai stratified split. Hal ini penting agar model belajar dan diuji pada proporsi kelas yang adil.

### Step 9 - Preprocessing dan Model AdaBoost
Cell ini membuat pipeline. Data kategorik diubah dengan One-Hot Encoding, lalu model AdaBoost disiapkan.

```python
try:
    encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
except TypeError:
    encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', encoder, categorical_cols),
        ('num', 'passthrough', numeric_cols)
    ]
)

# AdaBoost dengan Decision Tree sederhana sebagai weak learner.
try:
    adaboost_model = AdaBoostClassifier(
        estimator=DecisionTreeClassifier(max_depth=1, random_state=42),
        n_estimators=100,
        learning_rate=1.0,
        random_state=42
    )
except TypeError:
    adaboost_model = AdaBoostClassifier(
        base_estimator=DecisionTreeClassifier(max_depth=1, random_state=42),
        n_estimators=100,
        learning_rate=1.0,
        random_state=42
    )

model_pipeline = Pipeline(
    steps=[
        ('preprocess', preprocessor),
        ('model', adaboost_model)
    ]
)

print("Pipeline preprocessing dan model AdaBoost berhasil dibuat.")
```

### Penjelasan Output
Output `Pipeline preprocessing dan model AdaBoost berhasil dibuat` berarti alur pemrosesan data dan model sudah disusun. Pipeline ini menggabungkan One Hot Encoding untuk data kategorik, passthrough untuk data numerik, dan algoritma AdaBoost sebagai model klasifikasi. Pipeline membuat proses training lebih rapi karena preprocessing dan model berjalan dalam satu alur.

### Step 10 - Melatih Model
Cell ini melatih model AdaBoost menggunakan data training.

```python
model_pipeline.fit(X_train, y_train)

print("Model AdaBoost berhasil dilatih.")
```

### Penjelasan Output
Output `Model AdaBoost berhasil dilatih` berarti model sudah belajar dari data training. Pada tahap ini, AdaBoost membuat beberapa decision stump atau pohon keputusan sederhana, lalu menggabungkannya menjadi model yang lebih kuat. Data testing belum dipakai pada proses training, sehingga evaluasi tetap lebih objektif.

### Step 11 - Prediksi Data Testing
Cell ini memakai model untuk memprediksi data testing dan menampilkan contoh hasil prediksi.

```python
y_pred = model_pipeline.predict(X_test)
y_proba = model_pipeline.predict_proba(X_test)[:, 1]

hasil_prediksi = pd.DataFrame({
    'Aktual': y_test.values,
    'Prediksi': y_pred,
    'Probabilitas_High_Yield': y_proba
})

hasil_prediksi['Aktual_Label'] = hasil_prediksi['Aktual'].map({
    0: 'Low Yield',
    1: 'High Yield'
})

hasil_prediksi['Prediksi_Label'] = hasil_prediksi['Prediksi'].map({
    0: 'Low Yield',
    1: 'High Yield'
})

display(hasil_prediksi.head(10))
```

### Penjelasan Output
Output menampilkan contoh hasil prediksi pada data testing. Kolom `Aktual` menunjukkan kelas asli, sedangkan `Prediksi` menunjukkan hasil tebakan model. Kolom `Probabilitas_High_Yield` menunjukkan peluang suatu data masuk kelas High Yield. Jika probabilitas mendekati 1, model lebih yakin data tersebut High Yield. Jika mendekati 0, model lebih yakin data tersebut Low Yield.

### Step 12 - Evaluasi Model Utama
Cell ini menghitung accuracy, precision, recall, F1-score, ROC-AUC, classification report, dan confusion matrix.

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label=1)
recall = recall_score(y_test, y_pred, pos_label=1)
f1 = f1_score(y_test, y_pred, pos_label=1)
roc_auc = roc_auc_score(y_test, y_proba)

print("HASIL EVALUASI MODEL ADABOOST")
print("=" * 50)
print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1-Score  : {f1:.4f}")
print(f"ROC-AUC   : {roc_auc:.4f}")

print("\nClassification Report:")
print(classification_report(
    y_test,
    y_pred,
    labels=[0, 1],
    target_names=['Low Yield', 'High Yield']
))

cm = confusion_matrix(y_test, y_pred, labels=[0, 1])
print("Confusion Matrix:")
print(cm)
```

### Penjelasan Output
Output menunjukkan performa model utama AdaBoost pada data testing. Nilai **accuracy 0,8974** berarti sekitar **89,74%** data testing berhasil diklasifikasikan dengan benar. Nilai **precision 0,8811** berarti prediksi High Yield cukup tepat. Nilai **recall 0,9188** berarti model mampu menangkap sebagian besar data High Yield. Nilai **F1-score 0,8996** menunjukkan keseimbangan precision dan recall cukup baik. Nilai **ROC-AUC 0,9789** menunjukkan kemampuan model membedakan Low Yield dan High Yield sangat kuat.

### Step 13 - Visualisasi Confusion Matrix
Cell ini menampilkan confusion matrix dalam bentuk heatmap agar kesalahan prediksi lebih mudah dibaca.

```python
plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Low Yield', 'High Yield'],
    yticklabels=['Low Yield', 'High Yield']
)
plt.title('Confusion Matrix - AdaBoost')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.tight_layout()
plt.show()
```

### Penjelasan Output
Output berupa confusion matrix. Angka diagonal menunjukkan prediksi benar. Model berhasil memprediksi **205 data Low Yield** dan **215 data High Yield** dengan benar. Kesalahan model terdiri dari **29 data Low Yield** yang diprediksi High Yield dan **19 data High Yield** yang diprediksi Low Yield. Semakin besar angka diagonal dibanding angka di luar diagonal, semakin baik performa model.

### Step 14 - Cek Overfitting Sederhana
Cell ini membandingkan akurasi data training dan testing. Selisih kecil menunjukkan indikasi overfitting tidak kuat.

```python
train_pred = model_pipeline.predict(X_train)
test_pred = model_pipeline.predict(X_test)

train_accuracy = accuracy_score(y_train, train_pred)
test_accuracy = accuracy_score(y_test, test_pred)

print("Accuracy training:", round(train_accuracy, 4))
print("Accuracy testing :", round(test_accuracy, 4))
print("Selisih accuracy :", round(abs(train_accuracy - test_accuracy), 4))

if abs(train_accuracy - test_accuracy) <= 0.05:
    print("Interpretasi: selisih kecil, indikasi overfitting tidak kuat.")
else:
    print("Interpretasi: selisih cukup besar, perlu cek potensi overfitting.")
```

### Penjelasan Output
Output membandingkan accuracy training dan testing. Accuracy training sebesar **0,8957**, sedangkan accuracy testing sebesar **0,8974**. Selisihnya hanya **0,0017**. Selisih kecil ini menunjukkan model tidak menunjukkan indikasi overfitting yang kuat. Artinya, model tidak hanya menghafal data training, tetapi juga masih mampu bekerja baik pada data testing.

### Step 15
Cell ini melakukan 5-fold cross validation untuk melihat kestabilan model pada beberapa pembagian data.

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores = cross_val_score(
    model_pipeline,
    X,
    y,
    cv=cv,
    scoring='accuracy'
)

print("Hasil accuracy tiap fold:")
print(cv_scores)

print("Rata-rata accuracy:", round(cv_scores.mean(), 4))
print("Standar deviasi   :", round(cv_scores.std(), 4))
```

### Penjelasan Output
Output cross validation menunjukkan accuracy pada 5 fold, yaitu beberapa hasil pengujian dari pembagian data yang berbeda. Rata-rata accuracy adalah **0,8854** dengan standar deviasi **0,0167**. Nilai ini menunjukkan performa model cukup stabil. Standar deviasi yang kecil berarti hasil antar fold tidak berbeda jauh.

### Step 16 - Feature Importance Model Utama

### Step 20 - Perbandingan Model dengan dan tanpa Normalisasi
Cell ini membandingkan performa AdaBoost tanpa normalisasi, AdaBoost dengan normalisasi, dan AdaBoost 6 fitur.

```python
try:
    encoder_scaler = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
except TypeError:
    encoder_scaler = OneHotEncoder(handle_unknown='ignore', sparse=False)

preprocessor_scaler = ColumnTransformer(
    transformers=[
        ('cat', encoder_scaler, categorical_cols),
        ('num', StandardScaler(), numeric_cols)
    ]
)

try:
    adaboost_scaler = AdaBoostClassifier(
        estimator=DecisionTreeClassifier(max_depth=1, random_state=42),
        n_estimators=100,
        learning_rate=1.0,
        random_state=42
    )
except TypeError:
    adaboost_scaler = AdaBoostClassifier(
        base_estimator=DecisionTreeClassifier(max_depth=1, random_state=42),
        n_estimators=100,
        learning_rate=1.0,
        random_state=42
    )

pipeline_scaler = Pipeline(
    steps=[
        ('preprocess', preprocessor_scaler),
        ('model', adaboost_scaler)
    ]
)

pipeline_scaler.fit(X_train, y_train)

y_pred_scaler = pipeline_scaler.predict(X_test)
y_proba_scaler = pipeline_scaler.predict_proba(X_test)[:, 1]

accuracy_scaler = accuracy_score(y_test, y_pred_scaler)
roc_auc_scaler = roc_auc_score(y_test, y_proba_scaler)

comparison_df = pd.DataFrame({
    'Model': ['AdaBoost Tanpa Normalisasi', 'AdaBoost Dengan Normalisasi', 'AdaBoost 6 Fitur'],
    'Accuracy': [accuracy, accuracy_scaler, accuracy_6],
    'Precision_High_Yield': [precision, precision_score(y_test, y_pred_scaler, pos_label=1), precision_6],
    'Recall_High_Yield': [recall, recall_score(y_test, y_pred_scaler, pos_label=1), recall_6],
    'F1_High_Yield': [f1, f1_score(y_test, y_pred_scaler, pos_label=1), f1_6],
    'ROC_AUC': [roc_auc, roc_auc_scaler, roc_auc_6]
})

display(comparison_df)
```

### Penjelasan Output
Output membandingkan tiga model. Model AdaBoost tanpa normalisasi dan dengan normalisasi menghasilkan accuracy yang sama, yaitu **0,8974**. Ini terjadi karena AdaBoost berbasis decision tree tidak terlalu sensitif terhadap skala data. Model **AdaBoost 6 Fitur** memiliki accuracy tertinggi, yaitu **0,9124**, dengan F1 High Yield **0,9194**. Maka, pada dataset ini, model dengan 6 fitur lebih efektif dibanding model dengan semua fitur.
